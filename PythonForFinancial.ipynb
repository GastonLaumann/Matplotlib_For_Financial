```python
#Two dimensional sample data set
y = np.random.standard_normal((20,2)).cumsum(axis=0)

y
```

```python
plt.figure(figsize=(10,6))  
plt.plot(y[:,0], lw=1.5, label="first")   
plt.plot(y[:,1], lw=1.5, label="second")   
plt.plot(y, "ro")
plt.legend(loc=0)
plt.xlabel("index")   
plt.ylabel("value")   
plt.title("A Simple Plot")
```

```python
#Rescales the first data set
y[:,0]=y[:,0]*100

plt.figure(figsize=(10,6))  
plt.plot(y[:,0], lw=1.5, label="first")   
plt.plot(y[:,1], lw=1.5, label="second")   
plt.plot(y, "ro")
plt.legend(loc=0)
plt.xlabel("index")   
plt.ylabel("value")   
plt.title("A Simple Plot")
```

```python
# Introduces a second y-axis to the plot

fig, ax1 = plt.subplots()
plt.plot(y[:,0], "b", lw=1.5, label="first")
plt.plot(y[:,0], "ro")
plt.legend(loc=8)
plt.xlabel("index")
plt.ylabel("value first")
plt.title("A Simple Plot")
ax2 = ax1.twinx()
plt.plot(y[:,1], "g", lw=1.5, label="second")
plt.plot(y[:,1], "ro")
plt.legend(loc=0)
plt.ylabel("value second")
```

```python
# Two separate subplots
plt.figure(figsize=(10,6))
plt.subplot(2,2,1)                       #Two rows, one column, plot #1
plt.plot(y[:,0], lw=1.5, label="first")
plt.plot(y[:,0], "ro")
plt.legend(loc=0)
plt.ylabel("value")
plt.title("A Simple Plot")

plt.subplot(2,2,2)                         #Two rows, one column, plot #2
plt.plot(y[:,1], "g", lw=1.5, label="second")
plt.plot(y[:,1], "ro")
plt.legend(loc=0)
plt.ylabel("value")

plt.subplot(2,2,3)                       #Two rows, one column, plot #3
plt.plot(y[:,0], lw=1.5, label="first")
plt.plot(y[:,0], "ro")
plt.legend(loc=0)
plt.ylabel("value")
plt.title("A Simple Plot")

plt.subplot(2,2,4)                         #Two rows, one column, plot #4
plt.bar(np.arange(len(y)), y[:,1], width=0.5, color="g", label ="Third")
plt.legend(loc=0)
plt.xlabel("index")
plt.title("Third Data Set")
```

```python
#Mathematically plot

def func(x):
  return 0.5 * np.exp(x) + 1          # Function definition

a,b = 0.5, 1.5                        # The integral limits
x = np.linspace(0,2)                  # The X values to plot de function
y = func(x)                           # The Y values to plot the function
Ix = np.linspace(a,b)                 # The X values within the integral limits
Iy = func(Ix)                         # The Y values within the integral limits
verts = [(a,0)] + list(zip(Ix,Iy)) + [(b,0)]    # The list object with tuples representing coordinates to be plotted.
```

```python
from matplotlib.patches import Polygon

fig, ax = plt.subplots(figsize=(10,6))
plt.plot(x,y,"b", linewidth=2)
```

```python
fig, ax = plt.subplots(figsize=(10,6))
plt.plot(x,y,"b", linewidth=2)

plt.ylim(bottom=0)                  # set the Y axis
poly = Polygon(verts, facecolor="0.7", edgecolor="0.5")   # Define the limits of the polygon
ax.add_patch(poly)                                        # Plot the polygon
```

```python
fig, ax = plt.subplots(figsize=(10,6))
plt.plot(x,y,"b", linewidth=2)

plt.ylim(bottom=0)                  # set the Y axis
poly = Polygon(verts, facecolor="0.7", edgecolor="0.5")   # Define the limits of the polygon
ax.add_patch(poly)                                        # Plot the polygon

plt.text(0.5 * (a + b), 1, r'$\int_a^b f(x)\mathrm{d}x$',
horizontalalignment='center', fontsize=20)                # Plot the Integral Function 
plt.figtext(0.9,0.075, '$x$')
plt.figtext(0.075, 0.9, '$f(x)$')                         # Places the axis labels

ax.set_xticks((a, b))
ax.set_xticklabels(('$a$', '$b$'))                        # Places the x labels
ax.set_yticks([func(a), func(b)])                         # Places the y labels
```
